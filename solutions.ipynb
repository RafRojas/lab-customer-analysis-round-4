```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')

display(df)


#print('This are the following types for each column: ',(df.dtypes))
```

```python
# Sort by category

df_num = df.select_dtypes(include=np.number)

df_cat = df.select_dtypes(include=object)

list(df_num)
```

```python
# Use seaborn library to construct distribution plots for the numerical variables:

for i in df_num:
    sns.displot(df_num[i])
    plt.show
```

```python
# Use Matplotlib to construct histograms

df['Customer Lifetime Value'].hist(bins=100)
plt.title('Customer Lifetime Value')
plt.show()

df['Income'].hist()
plt.title('Income')
plt.show()

df['Monthly Premium Auto'].hist()
plt.title('Monthly Premium Auto')
plt.show()

df['Months Since Last Claim'].hist()
plt.title('Months Since Last Claim')
plt.show()

df['Months Since Policy Inception'].hist()
plt.title('Months Since Policy Inception')
plt.show()

df['Number of Open Complaints'].hist()
plt.title('Number of Open Complaints')
plt.show()

df['Number of Policies'].hist()
plt.title('Number of Policies')
plt.show()

df['Total Claim Amount'].hist()
plt.title('Total Claim Amount')
plt.show

# Do the distributions for different numerical variables look like a normal distribution

print('Non of the histograms follow the normal distribution')
```

```python
# For the numerical variables, check the multicollinearity between the features.

correlations_matrix = df_num.corr()
sns.heatmap(round(correlations_matrix, 2), annot=True) 
plt.show()
```
